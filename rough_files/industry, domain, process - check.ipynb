```python
for industry in res:
    if len(res[industry]) != 10:
        print ("No. of domains in {} industry : {}".format(industry,len(res[industry])))
```

```python
for industry in res:
    for domain in res[industry]:
        if len(res[industry][domain]) != 10:
            print ("No. of processes in {} industry and {} domain : {}".format(industry,domain,len(res[industry][domain])))
```
