```python
def compare_dicts(dict1, dict2):
    if type(dict1) != type(dict2):
        return False

    if isinstance(dict1, dict):
        if set(dict1.keys()) != set(dict2.keys()):
            return False

        for key in dict1:
            if not compare_dicts(dict1[key], dict2[key]):
                return False

    elif isinstance(dict1, list):
        if sorted(dict1) != sorted(dict2):
            return False

    else:
        return dict1 == dict2

    return True

result = compare_dicts(dict1, dict2)

if result:
    print("The dictionaries are equal.")
else:
    print("The dictionaries are not equal.")
```

```python
def compare_dicts(dict1, dict2, current_key=""):
    differing_keys = []

    if type(dict1) != type(dict2):
        return [current_key]

    if isinstance(dict1, dict):
        if set(dict1.keys()) != set(dict2.keys()):
            differing_keys.append(current_key)

        for key in dict1:
            new_key = f"{current_key}.{key}" if current_key else key
            differing_keys.extend(compare_dicts(dict1[key], dict2.get(key, None), new_key))

    elif isinstance(dict1, list):
        if sorted(dict1) != sorted(dict2):
            differing_keys.append(current_key)

    elif dict1 != dict2:
        differing_keys.append(current_key)

    return differing_keys

differing_keys = compare_dicts(dict1, dict2)

if not differing_keys:
    print("The dictionaries are equal.")
else:
    print("The dictionaries are not equal. Differing keys:")
    for key in differing_keys:
        print(f"- {key}")
```
